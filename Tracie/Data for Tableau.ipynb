```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cpi
```

```python
mojo_upd_df = pd.read_csv('Mojo_budget_update.csv')
```

```python
mojo_upd_df = mojo_upd_df.loc[pd.notnull(mojo_upd_df["release_date"])].reset_index(drop=True)
```

```python
mojo_upd_df = mojo_upd_df.loc[pd.notnull(mojo_upd_df["mpaa"])].reset_index(drop=True)
```

```python
mojo_upd_df = mojo_upd_df.drop(["distributor","writer","producer","composer","cinematographer","international","genre_2","genre_3","genre_4","trivia","html"],axis=1)
```

```python
mojo_upd_df = mojo_upd_df.loc[pd.notnull(mojo_upd_df["domestic"])].reset_index(drop=True)
```

```python
mojo_upd_df = mojo_upd_df.loc[pd.notnull(mojo_upd_df["worldwide"])].reset_index(drop=True)
```

```python
mojo_upd_df = mojo_upd_df.loc[pd.notnull(mojo_upd_df["main_actor_4"])].reset_index(drop=True)
```

```python
#Convert run_time in to minutes
times = []

for index,row in mojo_upd_df.iterrows():
    vals = row.run_time.split(" ")
    if len(vals)>2:
        run_time = int(vals[0])*60 + int(vals[2])    
    else:
        run_time = int(vals[0])*60
    times.append(run_time)

mojo_upd_df["runtime_conv"] = times    
```

```python
#Combining release date and year

full_date = []

for index,row in mojo_upd_df.iterrows():
    vals = str(row["release_date"]) + " " + str(row["year"])
    full_date.append(vals)

mojo_upd_df["full_date"] = full_date   
mojo_upd_df["full_date"] = pd.to_datetime(mojo_upd_df["full_date"])                    
mojo_upd_df.head()
```

```python
mojo_upd_df = mojo_upd_df.drop(["run_time","release_date"],axis=1)
```

```python
mojo_upd_df["infl_budget"] = mojo_upd_df.apply(lambda x: cpi.inflate(x["budget"], x["year"]), axis =1)
```

```python
mojo_upd_df["infl_domestic"] = mojo_upd_df.apply(lambda x: cpi.inflate(x["domestic"], x["year"]), axis =1)
```

```python
mojo_upd_df["infl_worldwide"] = mojo_upd_df.apply(lambda x: cpi.inflate(x["worldwide"], x["year"]), axis =1)
```

```python
mojo_upd_df = mojo_upd_df.drop(["budget","domestic","worldwide"],axis=1)
```
