# Gather data and Data Cleaning

```python
rotten_tomatoes_df = pd.read_csv("Resources/rotten_tomatoes_movies.csv")
```

```python
rotten_tomatoes_df.info()
```

```python
df = rotten_tomatoes_df.copy()
```

```python
df = df.loc[:,["rotten_tomatoes_link","movie_title", "content_rating", "genres", "directors","actors","original_release_date", "runtime", "production_company","tomatometer_status","tomatometer_rating","audience_rating","tomatometer_top_critics_count","tomatometer_fresh_critics_count","tomatometer_rotten_critics_count"]]
df = df.dropna().reset_index(drop=True)

df.info()
```

```python
df.head()
```

```python
df["original_release_date"]=pd.to_datetime(df.original_release_date)
df.info()
```

```python
df["year"]=df.original_release_date.apply(lambda x: x.year)
```

```python
df["month"]=df.original_release_date.apply(lambda x: x.month)
```

```python
df.head()
```

```python

df.to_csv("Resources/cleaned_rotten_tomatoes_tableau.csv",index=False)
cleaned_rotten_tomatoes = pd.read_csv("Resources/cleaned_rotten_tomatoes_tableau.csv")
cleaned_rotten_tomatoes.head() 
```
