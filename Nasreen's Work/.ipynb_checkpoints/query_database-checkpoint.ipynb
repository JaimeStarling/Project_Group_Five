```python
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "Resources/cleaned_mojo_budget_upd.csv"
clean_mojo_upd_df = pd.read_csv(data)
clean_mojo_upd_df.head()
```

```python
clean_mojo_upd_df.info()
```

```python
movies_df = clean_mojo_upd_df.filter(['movie_id','title','director','main_actor_1','main_actor_2','main_actor_3','main_actor_4','full_date'], axis=1)
movies_df.info()
```

```python
movies_df.head()
```

```python
revenue_df = clean_mojo_upd_df.filter(['movie_id','title','budget','domestic'], axis=1)
revenue_df.info()
```

```python
revenue_df.head()
```

```python
genre_df = clean_mojo_upd_df.filter(['movie_id','title','mpaa','genre_1','runtime_conv'], axis=1)
genre_df.info()
```

```python
genre_df.head()
```

```python
genre_df.info()
```

```python
database_path = "mojo_update.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
sql_movies = """CREATE TABLE IF NOT EXISTS movies(
    movie_id varchar(100) PRIMARY KEY,
    title varchar(100),
    director varchar(100) ,
    main_actor_1 varchar(500),
    main_actor_2 varchar(500),
   main_actor_3 varchar(500),
    main_actor_4 varchar(500),
    full_date varchar(100)
    )"""
```

```python
engine.execute(sql_movies)
```

```python
sql_revenue = """CREATE TABLE IF NOT EXISTS revenue_details(
    movie_id varchar(100) PRIMARY KEY,
    title varchar(100),
    budget double ,
    domestic double
    )"""
```

```python
engine.execute(sql_revenue)
```

```python
sql_genre = """CREATE TABLE IF NOT EXISTS genre_details(
    movie_id varchar(100)PRIMARY KEY,
    title varchar(100),
    mpaa varchar(100) ,
    genre_1 varchar(500),
    runtime_conv int
    )"""
```

```python
engine.execute(sql_genre)
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
movies_df.to_sql("movies", con=engine, method="multi", index=False, if_exists="replace")
```

```python
# Make a connection to the SQL database
conn = engine.connect()

query = """
        SELECT
            *
        FROM
            movies;
        """

df1 = pd.read_sql(query, conn)
df1.head()
```

```python
# Make a connection to the SQL database and display the movie details greater than full_date 2000-11-1
conn = engine.connect()

query = """
        SELECT
            *
        FROM
            movies
            where (full_date > "2000-11-1");
        """

df2 = pd.read_sql(query, conn)
df2.head()
```

```python
revenue_df.to_sql("revenue_details", con=engine, method="multi", index=False, if_exists="replace")
```

```python
# Make a connection to the SQL database
conn = engine.connect()

query = """
        SELECT
            *
        FROM
            revenue_details;
        """

df3 = pd.read_sql(query, conn)
df3.head()
```

```python
# Make a connection to the SQL database
conn = engine.connect()

query = """
        SELECT
           max(budget),title
        FROM
            revenue_details;
        """

df4 = pd.read_sql(query, conn)
df4.head()
```

```python
genre_df.to_sql("genre_details", con=engine, method="multi", index=False, if_exists="replace")
```

```python
# Make a connection to the SQL database
conn = engine.connect()

query = """
        SELECT
            *
        FROM
            genre_details;
        """

df5 = pd.read_sql(query, conn)
df5.head()
```

```python
# Make a connection to the SQL database for displaying the count of movies based on genre
conn = engine.connect()

query = """
        SELECT
         count(genre_1),genre_1
        FROM
            genre_details
            group by genre_1 ;
        """

df6 = pd.read_sql(query, conn)
df6.head()
```
