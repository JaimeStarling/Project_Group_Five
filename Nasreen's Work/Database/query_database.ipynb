```python
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
database_path = "mojo_update.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
# Make a connection to the SQL database for displaying the count of movies based on genre
conn = engine.connect()

query = """
        SELECT
         * from
        movies_data;
        """

df1 = pd.read_sql(query, conn)
df1.head()
```

```python
conn = engine.connect()
genre = '"Drama"'
release_date = "2019-12-01"
min_budget = 100000
max_budget = 600000
query = f"""
    SELECT 
    m.title,
    m.main_actor_1,
    m.main_actor_2,
    d.director,
    g.genre_1,
    m.release_date,
    m.budget
    FROM 
   movies_data as m
    INNER JOIN genre_details as g ON m.genre_id = g.genre_id
    INNER JOIN director_details as d ON m.director_id = d.director_id
    where 
    g.genre_1 in ({genre}) and 
    m.release_date >= ({release_date}) and 
    m.budget >= {min_budget} and m.budget <= {max_budget}
    order by m.title; 
   
    """
df14 = pd.read_sql(query, conn)
df14
```

```python
conn = engine.connect()
genre = '"Drama"'
release_date = "2019-12-01"
min_budget = 100000
max_budget = 600000
mpaa = '"R"'
query = f"""
    SELECT 
    m.movie_id,m.title,mp.mpaa,d.director,m.main_actor_1,m.main_actor_2,m.main_actor_3,m.main_actor_4,m.domestic,m.budget,
    g.genre_1,m.runtime,m.release_date
    FROM 
   movies_data as m
    INNER JOIN genre_details as g ON m.genre_id = g.genre_id
    INNER JOIN director_details as d ON m.director_id = d.director_id
    INNER JOIN mpaa_details as mp ON m.mpaa_id = mp.mpaa_id
    
    where 
    g.genre_1 in ({genre}) and 
    m.release_date >= ({release_date}) and 
    m.budget >= {min_budget} and m.budget <= {max_budget} and mp.mpaa in ({mpaa})
    order by m.title; 
   
    """
df15 = pd.read_sql(query, conn)
df15
```
