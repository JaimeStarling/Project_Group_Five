```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "../Resources/cleaned_mojo_budget_upd.csv"
clean_mojo_upd_df = pd.read_csv(data)
clean_mojo_upd_df.head()
```

```python
genre_df = clean_mojo_upd_df.drop_duplicates(subset=["genre_1"], keep='first')
genre_df.head()
```

```python
genre_data = genre_df["genre_1"].reset_index(drop=True)
genre_data
```

```python
genre_data_df = pd.DataFrame({
    'genre_id':np.arange(1,genre_data.size+1,1),
    'genre_1':genre_data
})
```

```python
genre_data_df
```

```python
# genre_data_np =genre_data.to_numpy()
# genre_data_np
# genre_id = []
# # # genre_list = ['Comedy','Crime',"Drama","Action","Documentary","Fantasy","Adventure","Horror","Biography","Mystery","Animation","Family","Sci-Fi","Romance","Music"]
# # # genre_list = genre_data.genre_1
# for index,row in clean_mojo_upd_df.iterrows():
#     genre_list=np.where(genre_data_np == row.genre_1)[0]
#     genre_id.append(np.where(genre_data_np == row.genre_1)[0][0])
# clean_mojo_upd_df["genre_id"] = genre_id  

genre_id = []
for index,row in clean_mojo_upd_df.iterrows():
    indx_df = genre_data_df.loc[genre_data_df['genre_1'] == row.genre_1]
    indx = indx_df['genre_id'].iloc[0]
    genre_id.append(indx)

clean_mojo_upd_df["genre_id"] = genre_id       
clean_mojo_upd_df            
```

```python
director_df = clean_mojo_upd_df.drop_duplicates(subset=["director"], keep='first')
director_df.head()
```

```python
director_data = director_df["director"].reset_index(drop=True) 
director_data
```

```python
director_data_df = pd.DataFrame({
    'director_id':np.arange(1,director_data.size+1,1),
    'director':director_data
})
```

```python
director_data_df
```

```python
director_id = []
for index,row in clean_mojo_upd_df.iterrows():
    indx_df = director_data_df.loc[director_data_df['director'] == row.director]
    indx = indx_df['director_id'].iloc[0]
    director_id.append(indx)

clean_mojo_upd_df["director_id"] = director_id       
clean_mojo_upd_df    
```

```python
mpaa_df = clean_mojo_upd_df.drop_duplicates(subset=["mpaa"], keep='first')
mpaa_df
```

```python
mpaa_data = mpaa_df["mpaa"].reset_index(drop=True) 
mpaa_data
```

```python
mpaa_data_df = pd.DataFrame({
    'mpaa_id':np.arange(1,mpaa_data.size+1,1),
    'mpaa':mpaa_data
})
```

```python
mpaa_data_df
```

```python
mpaa_id = []
for index,row in clean_mojo_upd_df.iterrows():
    indx_df = mpaa_data_df.loc[mpaa_data_df['mpaa'] == row.mpaa]
    indx = indx_df['mpaa_id'].iloc[0]
    mpaa_id.append(indx)

clean_mojo_upd_df["mpaa_id"] = mpaa_id       
clean_mojo_upd_df    
```

```python
cols =['movie_id', 'title','mpaa_id','director_id','main_actor_1', 'main_actor_2',
       'main_actor_3', 'main_actor_4', 'budget', 'domestic','genre_id',
       'runtime', 'release_date']
movies_data_df = clean_mojo_upd_df.loc[:,cols]
```

```python
movies_data_df
```

```python
database_path = "mojo_update.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
sql_genre = """CREATE TABLE IF NOT EXISTS genre_details(
    genre_id varchar(100) PRIMARY KEY,
    genre_1 varchar(100)
    )"""
```

```python
engine.execute(sql_genre)
```

```python
sql_director = """CREATE TABLE IF NOT EXISTS director_details(
   director_id varchar(100)PRIMARY KEY,
    director varchar(100)
    )"""
```

```python
engine.execute(sql_director)
```

```python
sql_mpaa = """CREATE TABLE IF NOT EXISTS mpaa_details(
    mpaa_id varchar(100)PRIMARY KEY,
    mpaa varchar(100)
    )"""
```

```python
engine.execute(sql_mpaa)
```

```python
sql_movies = """CREATE TABLE IF NOT EXISTS movies_data(
    movie_id varchar(100) PRIMARY KEY,
    title varchar(100),
    mpaa_id varchar(100),
    director_id varchar(100),
    main_actor_1 varchar(500),
    main_actor_2 varchar(500),
   main_actor_3 varchar(500),
    main_actor_4 varchar(500),
    budget double,
    domestic double,
    genre_id varchar(100),
    runtime int,
    release_date varchar(100),
    FOREIGN KEY (mpaa_id) REFERENCES mpaa_details(mpaa_id),
    FOREIGN KEY (genre_id) REFERENCES genre_details(genre_id),
    FOREIGN KEY (director_id) REFERENCES director_details(director_id)
    )"""
```

```python
engine.execute(sql_movies)
```

```python
conn = engine.connect()
movie_title = '"Lionheart"'
query = f"""
    SELECT 
    m.title,
    mp.mpaa,
    m.main_actor_1,m.main_actor_2,m.budget,m.release_date,
    d.director,
    g.genre_1
    FROM 
   movies_data as m
    INNER JOIN mpaa_details as mp ON m.mpaa_id = mp.mpaa_id
    INNER JOIN genre_details as g ON m.genre_id = g.genre_id
    INNER JOIN director_details as d ON m.director_id = d.director_id
    where m.title in ({movie_title});
   
    """
df9 = pd.read_sql(query, conn)
df9
```

```python
conn = engine.connect()

query = """
    SELECT 
    m.title,
    m.main_actor_1,
    m.main_actor_2,
    d.director,
    g.genre_1,
    m.release_date
    FROM 
   movies_data as m
    INNER JOIN genre_details as g ON m.genre_id = g.genre_id
    INNER JOIN director_details as d ON m.director_id = d.director_id
    where release_date >= "2019-12-01"
    order by m.title;
   
    """
df13 = pd.read_sql(query, conn)
df13
```

```python
conn = engine.connect()
genre = '"Drama"'
release_date = "2019-12-01"
min_budget = 100000
max_budget = 600000
query = f"""
    SELECT 
    m.title,
    m.main_actor_1,
    m.main_actor_2,
    d.director,
    g.genre_1,
    m.release_date,
    m.budget
    FROM 
   movies_data as m
    INNER JOIN genre_details as g ON m.genre_id = g.genre_id
    INNER JOIN director_details as d ON m.director_id = d.director_id
    where 
    g.genre_1 in ({genre}) and 
    m.release_date >= ({release_date}) and 
    m.budget >= {min_budget} and m.budget <= {max_budget}
    order by m.title; 
   
    """
df14 = pd.read_sql(query, conn)
df14
```
